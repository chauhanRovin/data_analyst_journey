1. Creating a dictionary

```python
student = {"Name" : "Rovin Chauhan",
           "Age" : 28,
           "City" : "Ghaziabad"
           }

print(student)
```

2. Accessing Values

```python
print(student["Name"])
print(student["Age"])
```

3. Adding a New Key

```python
student["Target"] = "MANGO"
print(student)
```

4. Updating a Value

```python
student["City"] = "Mumbai"
print(student)
```

5. Loop Through a Dictionary

```python
for key, value in student.items():
  print(key, ":", value)
```
