Topic: Exception Handling (try, except, finally)

Real-world data is often messy. Programs crash if we don't handle errors properly. Exception handling helps your code continue running smoothly.

1. Basic Exception Handling

```python
try:
    num = int(input("Enter a number: "))
    print(100 / num)
except:
    print("Invalid input!")
```

2. Catch Specific Exceptions

```python
try:
    num = int(input("Enter a number: "))
    print(100 / num)

except ZeroDivisionError:
    print("You cannot divide by zero.")

except ValueError:
    print("Please enter a valid integer.")
```

3. Using else (else runs only if no exception occurs)

```python
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid input.")
else:
    print("You entered:", num)
```

4. Using finally (finally always executes, whether an error occurs or not)

```python
try:
    file = open("sample.txt", "r")
    print(file.read())
except FileNotFoundError:
    print("File not found.")
finally:
    print("Program finished.")
```

##Create a calculator that:

- Accepts two numbers
- Asks the user for an operator (+, -, *, /)
- Handles invalid operators
- Prevents division by zero
- Handles invalid numeric input

```python
num1 = int(input("Enter Ist number : "))
print("the Ist number is :", num1)

num2 = int(input("Enter IInd number"))
print("the IInd number is : ", num2)

operator = input("Enter operator :")

if (operator == "+"):
  print("The sum of numbers is :", num1 + num2)
elif(operator == "-"):
  print("The diff of numbers is :", num1 - num2)
elif(operator == "*"):
  print("The mul of numbers is :", num1 * num2)
elif(operator == "/"):
  try:
    print("The div of number is :", num1/num2)
  except ZeroDivisionError:
    print("You cannot divide by zero.")
else:
  print("Invalid operator")
```
