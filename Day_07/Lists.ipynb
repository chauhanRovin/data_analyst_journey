# Lists

Lists are one of the most important data structures in Python and are used extensively in data analysis.

1. Creating a List

```python
fruits = ["apple", "banana", "mango"]

print(fruits)
print(type(fruits))
```

2. Accessing Elements

```python
fruits = ["apple", "banana", "mango"]

print(fruits[0])
print(fruits[1])
print(fruits[-1])
```

3. Modifying a List

```python
fruits = ["apple", "banana", "mango"]

fruits[1] = "orange"

print(fruits)
```

4. Adding Elements

```python
fruits = ["apple", "banana", "orange"]

fruits.append("mango")

print(fruits)
```

5. Removing Elements

```python
fruits = ["apple", "banana", "mango"]

fruits.remove("banana")

print(fruits)
```

6. Loop Through a List

```python
fruits = ["apple", "banana", "mango"]

for fruit in fruits:
    print(fruit)
```
