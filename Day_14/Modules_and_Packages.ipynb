## Topic: Modules and Packages

One of Python's biggest strengths is its vast ecosystem of modules. Instead of writing everything from scratch, you can import existing functionality.

1. Import a Module

```python
import math

print(math.sqrt(25))
print(math.factorial(5))
print(math.pi)
```

2. Import Specific Functions

```python
from math import sqrt, factorial

print(sqrt(25))
print(factorial(6))
```

3. Use an Alias

```python
import math as m

print(m.pi)
print(m.sqrt(49))
```

4. Random Module

```python
import random as rn

print(rn.randint(1, 100))
```

5. Datetime Module

```python
import datetime as dt

print(dt.datetime.today())
```

Random Password Generator

```python
import random

characters = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789"

password = ""

for i in range(8):
    password += random.choice(characters)

print(password)
```
