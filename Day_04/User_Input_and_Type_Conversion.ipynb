## User Input

###Introduction
User input allows a program to accept data from the user while it is running. In Python, the input() function is used to take input from the keyboard.

Input always returns string

```python
age = int(input("Enter your age: "))

print("Next year you will be: ", age + 1)
```

##Type Conversion
###Introduction

Type conversion is the process of converting one data type into another. This is important because the input() function always returns data as a string.

```python
num1 = int(input("Enter the 1st number: "))
num2 = int(input("Enter the 2nd number: "))

sum = num1 + num2

print("Sum of number is: ", sum)
```

Practice 1: Sum of Two Numbers

```python
length = float(input("Enter the length: "))
breadth = float(input("Enter the breadth: "))

area = length * breadth

print("Area of rectangle is: ", area)
```
